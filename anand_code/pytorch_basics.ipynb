```python
import torch
import torch.nn.functional as F
```

```python
from torch.autograd import grad
```

```python
y = torch.tensor([1.0])
x1 = torch.tensor([1.1])
# Declare these w requires grad true so they're learnable through backwards pass
w1 = torch.tensor([2.2], requires_grad = True)
b = torch.tensor([0.0], requires_grad = True)
```

```python
z = x1*w1 + b
a = torch.sigmoid(z)
```

```python
loss = F.binary_cross_entropy(a,y)
```

```python
loss.backward()
```

```python
print(w1.grad)
print(b.grad)
```
